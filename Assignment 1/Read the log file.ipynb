# Read log.file in Pandas but in certain lines of the results of matches 

```python
log_file_path = 'Gaussian log.file/glucose.log'
# pattern1 = r'Mulliken atomic charges'  # keyword or other regular expression to find the targets
pattern2= r'E\(RHF\)' 
# the regular expression of \ can make the specific re as literal expression, in this case ''\(\)' make () as literal expression
pattern3= r'Low frequencies'
# pattern4 = r'Dipole moment '
pattern_list = [pattern2,pattern3]
log_data = [] # the content of sentenses constrained by the pattern wull be stored in this list

for pattern in pattern_list:
    file= open(log_file_path, 'r')
#   with open(log_file_path, 'r') as file: # with statement in this case is equal to such sth equal to a variable (file= open(log_file_path, 'r'))
#   In addition , with statement ensures that the file is properly closed after you finish working with it, even if an exception occurs.
    for line in file : # line=string is a default variable means show the whole content of the sentence
        match= re.search(pattern, line) # re.search(pattern, string)
            # (string,vaiable) in this case it is useful to search the big string, wherever its loc of the matched line
            # search from the top to bottom, but in the vi editor, the top is the bottom
        if match:
            log_data.append(line)

# data = {'Log Line': log_data} #{colume title : data list}
df= pd.DataFrame(log_data, columns=[r'E_tot & vibrational frequencies']) # columns=['index']
pd.set_option('display.max_colwidth', None)  # Setting the column width to unlimited
df
```

```python
log_file_path = 'Gaussian log.file/glucose.log'
pattern1= r'E\(RHF\)'
pattern2= r'Low frequencies'
pattern_list = [pattern1,pattern2]
log_data = []

for pattern in pattern_list:
    file= open(log_file_path, 'r') # is equivalent to  with open(log_file_path, 'r') as file: (have to remember the indent)
    for line in file :
        match= re.search(pattern, line)
        if match:
            log_data.append(line)

df= pd.DataFrame(log_data, columns=[r'E_tot & vibrational frequencies']) # columns=['index']
pd.set_option('display.max_colwidth', None)  # Setting the column width to unlimited
df
```

# Find the Dipole moment with regular expressions (convienient if master it)

```python
pattern_dipole = r'Dipole moment \(field-independent basis, Debye\):\n\s+X=\s+([\d.-]+)\s+Y=\s+([\d.-]+)\s+Z=\s+([\d.-]+)\s+Tot=\s+([\d.-]+)'

dipole_data = []

with open(log_file_path,'r') as file:
    for line in file :
        match= re.search(pattern_dipole,line) 
        if match:
            x= float(match.group(1)) # group(n) is defined on the pattern which is ([\d.-]+)
            y= float(match.group(2)) 
            z= float(match.group(3))
            tot= float(match.group(4))
            dipole_data.append(['x','y','z','tot'])
            break
df= pd.DataFrame(dipole_data, columns=['X','Y','Z','Tot'])
df
```
