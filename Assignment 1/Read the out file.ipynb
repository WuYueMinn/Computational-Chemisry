```python
import pandas as pd
import re
```

```python
out_file_path= 'psi4 out.file/glucose.out'
```

# Read Three lowest Frequency and Total Energy

```python
out_file_data=[]
pattern1 = r'Freq'
pattern2 = r'Total Energy = '
pattern_list = [pattern1, pattern2]

for pattern in pattern_list :
    with open(out_file_path, 'r') as file:  
        for line in file :
            match= re.search(pattern, line) 
            if match:
                out_file_data.append(line)
df= pd.DataFrame(out_file_data, columns=[r'E_tot & vibrational frequencies']) # columns=['index']
pd.set_option('display.max_colwidth', None)  # Setting the column width to unlimited
df.to_csv('csv of the out.file/E_tot & VF_glucose.csv',index=None)
df.head(24)
```

```python
df.tail(24)
```

# Read the Vibration Frequency

```python
pattern_VF=r'Vibration'
VF_data= []

with open(out_file_path, 'r') as file:
    for line in file:
        match = re.search(pattern_VF, line)
        if match:
            next_line = next(file).strip() # remove the blank of the sentence
            VF_data.append([match.group(), next_line]) 
            # match.group() show the match string(all content) , nunber in it relate to the (...) inside the srring, ex is last one cell

df = pd.DataFrame(VF_data, columns=['',''])
pd.set_option('display.max_colwidth', None) 
df.to_csv('csv of the log.file/VF_glucose.csv',index=None)
df
```

# Read the Dipole Moment

```python
pattern_dipole = r'Dipole Moment: \[D\]'

dipole_data = []

with open(out_file_path, 'r') as file:
    for line in file:
        match = re.search(pattern_dipole, line)
        if match:
            next_line = next(file).strip() # remove the blank of the sentence
            dipole_data.append([match.group(), next_line]) 
            # match.group() show the match string(all content) , nunber in it relate to the (...) inside the srring, ex is last one cell
df = pd.DataFrame(dipole_data, columns=['',''])
pd.set_option('display.max_colwidth', None) 
df.to_csv('csv of the log.file/DM_glucose.csv',index=None)
df.tail(15)
```

# Read the Mulliken Charges

```python
MAC = r'Mulliken Charges:'
MAC_data=[]
lines_per_row = 27
# the num of lines in each row of the displayed dataframe, if it too small, the next match (MAC) will jump to the next row (try =22)
with open(out_file_path,'r') as file:
    for line in file:
        match = re.search(MAC, line)
        if match:
            row_data = []
            #  the list of row_data represent the columns of displayed df 
            for _ in range(lines_per_row):
                row_data.append(line.strip())
                line = next(file).strip()
            MAC_data.append(row_data) 
            # the list of MAC_data is the num of the rows of displayed df 
column_atoms = [f'Atom {i+1}' for i in range(lines_per_row)]
columns_names = [''] + [''] + column_atoms[:24] +['']
df = pd.DataFrame(MAC_data, columns=columns_names)
df.to_csv('csv of the out.file/MAC_glucose.csv',index=None)
df
```
