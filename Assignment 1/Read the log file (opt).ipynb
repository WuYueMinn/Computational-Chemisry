```python
import re
import pandas as pd
```

# Clear look of Mulliken Charge, but not general

```python
log_file = 'Gaussian log.file/glucose.log'
Info_after_optimized = []
pattern1 = r'Normal termination'
pattern2 = r'SCF Done'
pattern3 = r'Dipole moment \(field-independent basis, Debye\):'
lines_after_p3 = 1
pattern4 = r'Mulliken atomic charges:'
lines_after_p4 = 25 # Adjust this value based on the number of atoms in your molecule
pattern5 = r'Frequencies'
pattern_list = [pattern2,pattern3 ,pattern4, pattern5]
found_termination = False # Suppose process of found termination hasn't done yet (flag False) 

with open(log_file, 'r') as file:
    for line in file:
    # It will iterate all the lines (how many times) in file until the end of the file. (Remember the process is straight forward from the first line to the last line )
        if not found_termination: # check if the process of found termination whether True or Not
            match_termination = re.search(pattern1, line)
            if match_termination: # The process is True, hence we flag the found termination True, such that the found termination be True to remaining code 
                found_termination = True
                
        else: # this will execute no matter the found_termination is True or not
            for pattern in pattern_list: # the iteration is more specific than the previous loop, and the process is also straight forward
#           Since its order is lower the previous loop and under the condition of whether found the termination pattern, it will iterate remaining lines after the termination pattern  
                match = re.search(pattern, line)
                if match:
                    Info_after_optimized.append(line.strip())
                    
                    if pattern == pattern3:
                        for _ in range(lines_after_p3):
                            line = next(file).strip()
                            Info_after_optimized.append(line)

                    if pattern == pattern4:
                        for _ in range(lines_after_p4):
                            line = next(file).strip()
                            Info_after_optimized.append(line)

    if not found_termination: # Correspond to the first if statement, if found_termination is True, then the following print won't proceed, otherwise print. 
        print('Optimization failed')

df = pd.DataFrame(Info_after_optimized,columns=['Information after Geometry Optimization'])
pd.set_option('display.max_colwidth', None)

df
```

# Mulliken Charge is a bit messy, but can cope with any log file 

```python
log_file = 'Gaussian log.file/glucose.log'
Info_after_optimized = []
pattern1 = r'Normal termination'
pattern2 = r'SCF Done'
pattern3 = r'Dipole moment \(field-independent basis, Debye\):'
lines_after_p3 = 1
pattern4 = r'Mulliken atomic charges:'
lines_after_p4 = 26 # Adjust this value based on the number of atoms in your molecule
pattern5 = r'Frequencies'
pattern_list = [pattern2,pattern3 ,pattern4, pattern5]
found_termination = False

with open(log_file, 'r') as file:
    for line in file:
        if not found_termination:
            match_termination = re.search(pattern1, line)
            if match_termination:
                found_termination = True
                
        else:
            for pattern in pattern_list:
                match = re.search(pattern, line)
                if match:
                    Info_after_optimized.append(line.strip())

                    if pattern == pattern3:
                        for _ in range(lines_after_p3):
                            line = next(file).strip()
                            Info_after_optimized.append(line)

                    if pattern == pattern4:
                        row_data= []
                        for _ in range(lines_after_p4):
                            line = next(file).strip()
                            row_data.append(line)
                        Info_after_optimized.append(row_data[1:])
    if not found_termination:
        print('Optimization failed')

df = pd.DataFrame(Info_after_optimized,columns=['Information after Geometry Optimization'])
pd.set_option('display.max_colwidth', None)

df.head(6)
```
